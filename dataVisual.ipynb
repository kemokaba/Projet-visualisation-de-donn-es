```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Chargement du données CSV
df = pd.read_csv('world-happiness-report-2021.csv')
```

```python
# sélectionner les colonnes dont nous avons besoin
df = df[['Country name', 'Logged GDP per capita', 'Ladder score','Regional indicator']]
```

```python
#trier les données par PIB par habitant par ordre croissant
df.sort_values(by='Logged GDP per capita', inplace=True)
```

```python
# créer un diagramme de dispersion
plt.scatter(df['Logged GDP per capita'], df['Ladder score'])

# ajouter des étiquettes d'axe et un titre
plt.xlabel('PIB par habitant')
plt.ylabel('Score de bonheur')
plt.title('Relation entre le PIB par habitant et le score de bonheur')
```

```python
# Créez un dictionnaire qui associe une région à une couleur
region_color = {
    'Western Europe': 'red',
    'North America and ANZ': 'green',
    'Middle East and North Africa': 'blue',
    'Latin America and Caribbean': 'orange',
    'Central and Eastern Europe': 'purple',
    'East Asia': 'yellow',
    'Southeast Asia': 'brown',
    'Commonwealth of Independent States': 'gray',
    'Sub-Saharan Africa': 'pink',
    'South Asia': 'cyan'
}

# Associer une région à une couleur.
df['color'] = df['Regional indicator'].map(region_color)

# create a scatter plot with colored data points
plt.scatter(df['Logged GDP per capita'], df['Ladder score'], c=df['color'])

# Ajouter une legende
handles = [plt.plot([],[],color=color, ls="", marker="o")[0] for color in region_color.values()]
labels = region_color.keys()
plt.legend(handles, labels, loc='lower right')

# Ajouter des étiquettes d'axe et un titre.
plt.xlabel('PIB par habitant')
plt.ylabel('Score de bonheur')
plt.title('Relation entre le PIB par habitant et le score de bonheur')
```
